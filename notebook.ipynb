You have been asked to work with a bank to clean the data they collected as part of a recent marketing campaign, which aimed to get customers to take out a personal loan. They plan to conduct more marketing campaigns going forward so would like you to ensure it conforms to the specific structure and data types that they specify so that they can then use the cleaned data you provide to set up a PostgreSQL database, which will store this campaign's data and allow data from future campaigns to be easily imported.

They have supplied you with a csv file called "bank_marketing.csv", which you will need to clean, reformat, and split the data, saving three final csv files. Specifically, the three files should have the names and contents as outlined below:

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/bank_marketing.csv')
```

```python
# client.csv file 
selected_columns = ['client_id', 'age', 'job', 'marital', 'education', 'credit_default', 'mortgage' ]
df['credit_default'] = df['credit_default'].map({'yes': 1, 'no': 0, 'unknown': 0}) 
df['credit_default'] = df['credit_default'].astype('boolean')
df['job'] = df['job'].str.replace('.', '_')
df['education'] = df['education'].str.replace('.', '_')
df['education'] = df['education'].replace('unknown', np.nan)
df['mortgage'] = df['mortgage'].map(lambda x:True if x == 'yes' else False)
df['mortgage'] = df['mortgage'].astype('boolean')
new_df = df[selected_columns]
new_df.to_csv('data/client.csv', index= False)
```

```python
# campagin.csv file
selected_columns = ['client_id', 'number_contacts', 'contact_duration', 'previous_campaign_contacts',
                     'campaign_outcome', 'previous_outcome','last_contact_date']
df['campaign_outcome'] = df['campaign_outcome'].map(lambda x:1 if x == 'yes' else 0) 
df['campaign_outcome'] = df['campaign_outcome'].astype('boolean')
df['previous_outcome'] = df['previous_outcome'].map(lambda x:1 if x == 'success' else 0)
df['previous_outcome'] = df['previous_outcome'].astype('boolean')

month_map = {
    'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6,
    'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12
}


df['month'] = df['month'].str.lower().map(month_map)
df['day'] = df['day'].astype('int')

# Step 3: Create the 'last_contact_date' column by combining 'day', 'month', and the fixed year (2022)
df['last_contact_date'] = df.apply(lambda row: f"2022-{row['month']:02d}-{row['day']:02d}", axis=1)

# Step 4: Convert 'last_contact_date' to datetime format (optional but recommended)
df['last_contact_date'] = pd.to_datetime(df['last_contact_date'], format='%Y-%m-%d')

new_df = df[selected_columns]
new_df.to_csv('data/campaign.csv', index= False)
```

```python
# economics.csv file 
selected_columns = ['client_id', 'cons_price_idx', 'euribor_three_months']
new_df = df[selected_columns]
new_df.to_csv('data/economics.csv', index= False)
new_df
```
